## Problem: 2

Generate an array of length $3n$ , filled with the cyclic pattern 1, 2, 3. Where $n$ is a natural number.

```python
# Write your code here

n = 3

length = 3 * n

cyclic_pattern = np.linspace(1,3,3 ,dtype="int")

cyc_len = len(cyclic_pattern)

while  cyc_len < length:
      cyclic_pattern = np.append(cyclic_pattern , [1 ,2 ,3])
      cyc_len = len(cyclic_pattern)


print(cyclic_pattern)
print("---")

# Solution2

x = np.tile([1 ,2 ,3] , n)

print(x)

```

## Problem: 4

Create a 10 x 10 arrays of zeros and then "frame" it with a border of ones.

## Problem: 5

Create an 8 x 8 array with a checkerboard pattern of zeros and ones.

```python
# Write your code here

even_index_row = np.tile([1 , 0] , 4)

odd_index_row = np.tile([0,1] , 4)


stacked_n = np.vstack((even_index_row , odd_index_row))

checkerboard = np.tile(stacked_n , (4 ,1))


print(checkerboard)
```

## Problem: 7 
Create a function which creates an $n×n$ array with $(i , j)$ -entry equal to $i+ j$

```python
# Write your code here

def new_array(n):
    i = np.arange(n).reshape(n,1)
    j = np.arange(n).reshape(1,n)
    return i + j

new_array(4)
```

# Problem: 8
Evaluate $cos$ and $sin$ on the interval $[0,1]$  with 1000 points and then stack the results into a tall array with rows being the $(cos(x),sin(x))$ entries.

```python
# Write your code here

interval = np.linspace(0 ,1, 1000)

cos_results = np.cos(interval).reshape(-1 ,1)
sin_results = np.sin(interval).reshape(-1 ,1)

tall_array = np.hstack((cos_results , sin_results))
print(tall_array)
```

# Problem: 9

Create a random $3×5$ array using the `np.random.rand(3 ,5)` function and compute: the sum of all the entries, the sum of the rows, and the sum of the columns.

```python
# Write your code here

random_array = np.random.rand(3 ,5)

sum_of_all = random_array.sum()

sum_of_rows = random_array.sum(axis=1)

sum_of_cols = random_array.sum(axis=0)



print(random_array)
print("---")
print(sum_of_all)
print("---")
print(sum_of_rows)
print("---")
print(sum_of_cols)
```
